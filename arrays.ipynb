# 1.find the maximum value in the array

```python
lst=[1,2,3,-4,5,6,-7,89,10,0,23,0,233,0]

index=lst[0]
for i in range(len(lst)):
    if lst[i]>index:
        index=lst[i]

print(index)  
```

# 2.find the minimum value in the array

```python
lst=[1,2,3,5,6,89,10,23,233]

index=lst[0]
for i in range(len(lst)):
    if lst[i]<index:
        index=lst[i]

print(index)
```

# 3.find the 2nd maximum value in the array
